```python
#Finding the value of integral using Monte Carlo
```

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def f(x):
    return 5*x**2 - 5*x**3
```

```python
N =10000
plt.plot(np.linspace(0,1,N), f(np.linspace(0,1,N)), 'r')
```

```python
#generate a large sample of points (x,y) in [0,1) and [0,1)
x = np.random.random(N)
y = np.random.random(N)

plt.scatter(x,y,s=1)
plt.plot(np.linspace(0,1,N), f(np.linspace(0,1,N)), 'r')

#fx is the value of the fuction evaluated at x 
fx = f(x)
```

```python
fx > y #true = 1, false = 0
```

```python
np.sum(fx > y)
```

```python
integral = np.sum(fx > y) / N 
integral
```
